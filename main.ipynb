```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("books-scrapy.csv")
df.head()
```

```python
df.columns
```

```python
df = df.drop(columns=["isbn10","isbn13","subTitle","title","publisherDate","title","category","url","depth","returnable"])
df.head()
```

```python
pd.DataFrame(df.author.value_counts())
```

```python
df.info()
```

```python
df.describe()
```

```python
df.height = df["height"].str.replace(r" mm", "").str.replace(r" ","")
df['height'] = pd.to_numeric(df['height'], errors='coerce').fillna(np.nan)
```

```python
df.price = df.price.str.replace(r"₹", "").str.replace(r",","").astype(float)
```

```python
df.weight = df.weight.str.replace(r" gr","")
df.width = df.width.str.replace(r" mm","")
df['weight'] = pd.to_numeric(df['weight'], errors='coerce').fillna(np.nan)
df['width'] = pd.to_numeric(df['width'], errors='coerce').fillna(np.nan)
```

```python
df['noOfPages'] = pd.to_numeric(df['noOfPages'], errors='coerce').fillna(np.nan)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.corr(numeric_only=True)
```

```python
df.isnull().sum()
```

```python
for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):
        df[col].fillna(df[col].median(),inplace=True)
    else:
        df[col].fillna(df[col].mode(),inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
object_cols = df.select_dtypes(include=['object'])
object_cols.columns
for col in object_cols.columns:
    df[col] = le.fit_transform(df[col])
df.head()
```

```python
df.corr()
```

```python
X = df.drop(columns=["price"])
y = df.price
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42,train_size=0.8)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_pred)
r2
```
